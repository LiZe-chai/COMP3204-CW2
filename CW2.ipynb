```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

## Task 1: Image Resizing

```python
image = cv2.imread('input.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
print("Image size (H, W) of original image is:", image.shape)
```

```python
resized_800x600 = cv2.resize(image,dsize=(800,600),interpolation=cv2.INTER_LANCZOS4)
print("Image size (H, W) after resizing is:", resized_800x600.shape)
```

```python
plt.imshow(resized_800x600)
```

```python
resized_half = cv2.resize(image,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_LANCZOS4)
print("Image size (H, W) after resizing is:", resized_half.shape)
```

```python
plt.imshow(resized_half)
```

## Task 2: Face Detection

```python
face = cv2.imread('face.jpg')
grayFace = cv2.cvtColor(face,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
faces = face_cascade.detectMultiScale(grayFace, scaleFactor=1.1, minNeighbors=10)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(face, (x, y), (x+w, y+h), (0, 0, 255), 2)
plt.imshow(cv2.cvtColor(face, cv2.COLOR_BGR2RGB))
```
